```python
from pycaret.regression import *
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/82108/OneDrive/바탕 화면/git/han_river_water_level/datafile.csv", index_col = 0)
pdf = pd.read_csv("C:/Users/82108/OneDrive/바탕 화면/git/han_river_water_level/awsfile.csv",index_col = 0)
```

```python
df1 = df.drop(columns = ["wl_1018680","wl_1018683","wl_1019630"])
df2 = df.drop(columns = ["wl_1018662","wl_1018683","wl_1019630"])
df3 = df.drop(columns = ["wl_1018680","wl_1018662","wl_1019630"])
df4 = df.drop(columns = ["wl_1018680","wl_1018683","wl_1018662"])
pdf1 = pdf.drop(columns = ["wl_1018680","wl_1018683","wl_1019630"])
pdf2 = pdf.drop(columns = ["wl_1018662","wl_1018683","wl_1019630"])
pdf3 = pdf.drop(columns = ["wl_1018680","wl_1018662","wl_1019630"])
pdf4 = pdf.drop(columns = ["wl_1018680","wl_1018683","wl_1018662"])
```

```python
reg_test_1 = setup(data= df1, target="wl_1018662", train_size=0.7, fold=5, session_id = 152, imputation_type='iterative', iterative_imputation_iters=10, categorical_iterative_imputer='lightgbm',silent = True)
```
